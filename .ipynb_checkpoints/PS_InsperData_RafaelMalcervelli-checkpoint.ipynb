## Adquirindo variáveis relevantes

```python
base = pd.read_csv("base_7.csv", delimiter=';')
base[:3]
```

```python
date = base['date']
listaDatas = []
for data in date:
    dia = str(data)[2:4]
    mes = str(data)[:2]
    ano = str(data)[4:]
    listaDatas.append(f"{dia}/{mes}/{ano}")

print(listaDatas[:5])
print("Os dados do database foram coletados apenas no mês de agosto, cujo está dentro da alta temporada de visitantes e turistas na cidade.")
```

```python
print(f"Colunas do dataframe: {list(base.columns)}")
base.describe()
```

```python
countAM = 0
countPM = 0
for obs in base['shift']:
    if obs == "AM":
        countAM +=1
    else:
        countPM += 1
        
print(f"Sessão foi observada {countAM} vezes durante a manhã e {countPM} vezes durante a tarde, notando uma singela maioria de observações realizadas durante a tarde")
```

```python
countYoung = 0
countAdult = 0
for age in base['age']:
    if age == "Adult":
        countAdult +=1
    else:
        countYoung += 1
print(f"O animal observado era da faixa etária Adulto em {countAdult} das observações, enquanto que os Jovens eram {countYoung}")
print(f"Nota-se que a porcentagem de adultos no parque em relação ao número total de observações foi de {((countAdult/3023)*100):.2f} %")
```

```python
countGray = 0
countCinnamon = 0
countBlack = 0
for color in base['primary_fur_color']:
    if color == "Gray":
        countGray +=1
    elif color == "Cinnamon":
        countCinnamon += 1
    else:
        countBlack += 1

print(f"A pelagem primária dos animais observados foi de {countGray} vezes a cor cinza, {countCinnamon} vezes da cor canela e {countBlack} da cor preta")
print(f"Logo, {((countGray/3023)*100):.2f} % dos esquilos apresentam a pelagem primária de cor cinza.\n")
print(f"É importante ressaltar que existem esquilos dessa éspecie que apresentam a pelagem preta, uma característica anormal e adquirida com a seleção natural. Os esquilos melanísticos parecem exibir uma tolerância ao frio mais alta do que os de pelagem cinza comum; quando exposto a -10   ° C, os esquilos pretos apresentaram uma redução de 18% na perda de calor, uma redução de 20% na taxa metabólica basal e um aumento de 11% na capacidade de termogênese sem tremores quando comparados aos de pelagem cinza comum. \n")
print(f"A partir de dados coletados, o clima do Central Park em épocas de inverno apresenta uma média de -10 graus Celsius, demonstrando um motivo para a aparição de tais colorações pretas dos esquilos.")
```

## Analisando o comportamento dos esquilos com relação a interação com humanos

```python
approachesTotal = base.loc[base["approaches"] == True].count()[0]
approachNotHappy = base.loc[base["approaches"] == True][base["kuks"] == False].count()[0]
approachHappy = base.loc[base["approaches"] == True][base["kuks"] == True].count()[0]
scared = base.loc[base["runs_from"] == True].count()[0]
indifferent = base.loc[base["indifferent"] == True].count()[0]

print(f"{((approachesTotal/3023)*100):.2f} % dos esquilos observados se aproximaram de algum humano em busca de comida.")
print(f"A partir das análises realizadas, apenas {((approachHappy/approachesTotal)*100):.2f} % dos esquilos que se aproximaram de um humano em busca de comida apresentaram um comportamento de felicidade nessa aproximação. Já {((approachNotHappy/approachesTotal)*100):.2f} % dos esquilos apenas se aproximaram dos humanos em busca de comida, sem nenhum tipo de sentimento envolvido. \n")
print(f"{((scared/3023)*100):.2f} % dos esquilos apresentou um comportamento de 'medo' em relação a aproximação de um ser humano. Tal porcentagem é relativamente baixa, visto que a maioria dos animais estão acostumados com a presença de humanos ja que seu habitat é o Central Park, um parque localizado em uma das maiores metrópoles do mundo, além de que {((indifferent/3023)*100):.2f} % dos animais foram indiferentes com relação a presença de um ser humano próximo. \n")
```

```python
# Realizando conversões necessárias para inserir latitude e longitude no mapa.
base_food = base.loc[base["approaches"] == True]

listLat = []
for lat in base_food['lat'].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '.').astype(float):
    listLat.append(float(lat))
    
listLong = []
for long in base_food['long'].str.replace('[A-Za-z]', '').str.replace('.', '').str.replace(',', '.').astype(float):
    listLong.append(float(long)) 
```

## Analisando dados em um mapa

```python
mapa = Map(center=(40.785091, -73.968285),zoom=14, basemap=basemaps.CartoDB.DarkMatter)

counter = 0
for lat in listLat:
    marker = Marker(location=[lat, listLong[counter]], draggable=False)
    mapa.add_layer(marker)
    counter += 1

#Times Square
mapa.add_layer(Marker(location=[40.758896, -73.985130], draggable=False, title="Times Square"))

#Museu de Arte Moderna
mapa.add_layer(Marker(location=[40.761509, -73.978271], draggable=False, title="Museu de Arte Moderna"))

#Empire State Building
mapa.add_layer(Marker(location=[40.748817, -73.985428], draggable=False, title="Empire State Building"))

#Top of the Rock - Rockfeller Building
mapa.add_layer(Marker(location=[40.759060, -73.979431], draggable=False, title="Top of the Rock - Rockfeller building"))

mapa
```

Acima pode-se observar que a maioria das aproximações dos esquilos em busca de comida foram realizadas ao sul do Central Park, local que mais se destaca em relação ao número de visitantes devido a sua proximidade com famosos pontos turísticos como a Times Square.
